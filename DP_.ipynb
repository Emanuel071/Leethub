```python
#memoization

class Solution(object):
    def fibonacci_recursive(self, n, dictionary_memoization):
        """
        :type n: int
        :rtype: int
        """
        if n in dictionary_memoization:
            return dictionary_memoization[n]
        if n <= 2:
            return 1
        dictionary_memoization[n] = Solution().fibonacci_recursive(n - 1,dictionary_memoization) + Solution().fibonacci_recursive(n - 2,dictionary_memoization)
        return dictionary_memoization[n]

dictionary = {}
Solution().fibonacci_recursive(100,dictionary)
```

## Traveler Grid

```python
class Solution(object):
    def gridTraveler(self, n, m,memo):
        """
        :type piles: List[int]
        :rtype: bool
        """
        key = str(m) + ',' + str(n)
        if key in memo:
            return memo[key]
        if n == 0 or m == 0:
            return 0
        if n == 1 and m == 1:
            return 1
        memo[key] = Solution().gridTraveler(n - 1, m,memo) + Solution().gridTraveler(n,m - 1,memo)
        return memo[key]

memo = {}
n = 18
m = 18
Solution().gridTraveler(n, m,memo)
```

## guide lines for solving dynamic probles
1. make it work (make a solution)
- visualize the problem as a tree
- implement the tree using recursion
- test it (big difference with code that is slow vs wrong)

2. make it efficient
- add a memo object 
- add a base case to return memo values
- return values into the memo

```python
class Solution(object):
    def canSum(self, targetSum, numbers,memo):
        """
        :type piles: List[int]
        :rtype: bool
        """
        if targetSum in memo:
            return memo[targetSum]        
        if targetSum == 0:
            return True
        if targetSum < 0:
            return False
        for value in numbers:
            remainder = targetSum - value
            if Solution().canSum(remainder, numbers,memo) == True:
                memo[targetSum] = True
                return True
        memo[targetSum] = False
        return False

memo = {}
numbers = [1,2,3,4]
targetSum = 100
Solution().canSum(targetSum, numbers,memo)
```
